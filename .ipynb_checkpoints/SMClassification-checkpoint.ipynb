## Speech / Music Classification

#### Classification of sound files into either speech or music using a KNN algorithm as well as an ANN and comparing the results of both implementations

## Importing the data

```python
from __future__ import print_function

import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display

from sklearn.neural_network import MLPClassifier
```

Get the file path to the included audio example

```python
speech_file = "./LibriSpeech/84-121123-0000.flac"
music_file = "./fma_wavs/000002.wav"
```

Load the audio as a waveform `y`
Store the sampling rate as `sr`

```python
y_speech, sr_speech = librosa.load(speech_file)
y_music, sr_music = librosa.load(music_file)
```

## Extracting and Plotting Features

Plot the waveform

```python
plt.figure(figsize=(15, 4))
plt.subplot(1,2,1)
arr = librosa.display.waveplot(y_speech, sr=sr_speech)
print(y.shape)
plt.title("Waveform of audiofile containing speech");

plt.subplot(1,2,2)
librosa.display.waveplot(y_music, sr=sr_music)
plt.title("Waveform of audiofile containing musc");
```

### Spectogram

```python
plt.figure(figsize=(15, 4))

FFT_speech = librosa.amplitude_to_db(np.abs(librosa.stft(y_speech)), ref=np.max)
plt.subplot(1,2,1)
librosa.display.specshow(FFT_speech, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram of speech file');
plt.xlabel('time');

FFT_music = librosa.amplitude_to_db(np.abs(librosa.stft(y_music)), ref=np.max)
plt.subplot(1,2,2)
librosa.display.specshow(FFT_music, y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram of music file');
plt.xlabel('time');
```

### Zero-Crossing Rate 


```python
plt.figure(figsize=(18, 6))

speech_crossrate = librosa.feature.zero_crossing_rate(y_speech)
music_crossrate  = librosa.feature.zero_crossing_rate(y_music)

plt.subplot(1,2,1)
plt.plot(speech_crossrate[0]);
plt.title('Zero crossing rate for frames of speech file');
plt.xlabel('frame');
plt.ylabel('zero-crossing rate');

plt.subplot(1,2,2)
plt.plot(music_crossrate[0]);
plt.title('Zero crossing rate for frames  of music file');
plt.xlabel('frame');
plt.ylabel('zero-crossing rate');
```

#### Variance of Zero Crossing Rate

```python
print('Variance of music file: ', np.var(librosa.feature.zero_crossing_rate(y_speech)))
print('Variance of speech file: ',np.var(librosa.feature.zero_crossing_rate(y_music)))
```

The variance of the zero-crossing rate for the speech file is an order of magnitude higher than for the music file

```python
import scipy.fftpack

# Number of samplepoints
N_music = int(y_music.shape[0])
N_speech = int(y_speech.shape[0])
# sample spacing
T = 1.0 / 800.0

speech_fft = scipy.fftpack.fft(y_speech)
music_fft = scipy.fftpack.fft(y_music)

speech_freqs = np.linspace(0.0, 1.0/(2.0*T), int(N_speech/2))

music_freqs = np.linspace(0.0, 1.0/(2.0*T), int(N_music/2))

plt.figure(figsize=(18, 6))

plt.subplot(1,2,1)
plt.plot(speech_freqs, 2.0/N_speech * np.abs(speech_fft[:N_speech//2]));

plt.subplot(1,2,2)
plt.plot(music_freqs, 2.0/N_music * np.abs(music_fft[:N_music//2]));
```

We can see that the fourier transform introduces many dimensions, but isn't a very unique feature for this application

```python
plt.figure(figsize=(18, 6))

plt.subplot(1,2,1)
mfccs_music = np.mean(librosa.feature.mfcc(y=y_music, sr=sr_music, n_mfcc=40).T,axis=0) 
plt.plot(mfccs_music)
plt.ylim(-60)

plt.subplot(1,2,2)
mfccs_speech = np.mean(librosa.feature.mfcc(y=y_speech, sr=sr_speech, n_mfcc=40).T,axis=0) 
plt.plot(mfccs_speech)
plt.ylim(-60)
```
